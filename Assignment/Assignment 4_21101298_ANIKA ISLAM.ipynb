```python
# task 1 
class Calculator :
    def __init__(self,num1,op,num2) :
        self.num1 = num1 
        self.op = op
        self.num2 = num2
        print('Let\'s Calculate!')
        print('Value1:',self.num1)
        print('Operator:',self.op)
        print('Value2:',self.num2)
    def add(self) :
        print('Result:',self.num1+self.num2)
    def sub(self) :
        print('Result:',self.num1-self.num2)
    def mul(self) :
        print('Result:',self.num1*self.num2)
    def div(self) :
        print('Result:',self.num1/self.num2)
        
#=================================================
num1 = int(input('enter 1st number: '))
op = input('enter operator : ')
num2 = int(input('enter 2nd number: '))
calc = Calculator(num1,op,num2)
if calc.op == '+' :
    calc.add()
elif calc.op == '-' :
    calc.sub()
elif calc.op == '*' :
    calc.mul()
elif calc.op == '/' :
    calc.div()
```

```python
# task 21 
class Doctor :
    def __init__(self,doc_id,doc_prof,doc_name,doc_spec) :
        self.doc_id = doc_id  
        self.doc_prof = doc_prof 
        self.doc_name = doc_name 
        self.doc_spec = doc_spec 
class Patient :
    def __init__(self,pat_id,pat_prof,pat_name,pat_age,pat_phone) :
        self.pat_id = pat_id 
        self.pat_prof = pat_prof
        self.pat_name = pat_name 
        self.pat_age = pat_age
        self.pat_phone = pat_phone 
class Hospital :
    def __init__(self,hospital_name) :
        self.hospital_name = hospital_name 
        self.num_doctor = 0 
        self.num_patient = 0 
        self.doctor_dict = {}
        self.patient_dict = {}
    def addDoctor(self,Doctor) :
        self.num_doctor+= 1
        self.doctor_id = Doctor.doc_id  
        self.doctor_name = Doctor.doc_name
        self.doctor_spec = Doctor.doc_spec
        if self.doctor_id not in self.doctor_dict :
            self.doctor_dict[self.doctor_id] = [self.doctor_name,self.doctor_spec]
    def getDoctorByID(self,d_id) :
        return 'Doctor\'s ID: {}\nName: {}\nSpeciality: {}'.format(self.doctor_id,self.doctor_name,self.doctor_spec)
    def addPatient(self,patient) :
        self.num_patient += 1 
        self.patient_id = patient.pat_id
        self.patient_name = patient.pat_name 
        self.patient_age = patient.pat_age 
        self.patient_phone = patient.pat_phone
        if self.patient_id not in self.patient_dict :
            self.patient_dict[self.patient_id] = [self.patient_name,self.patient_age,self.patient_phone]
    def getPatientByID(self,p_id) :
        return 'Doctor\'s ID: {}\nName: {}\nAge: {}\nPhone no.: {}'.format(self.patient_id,self.patient_name,self.patient_age,self.patient_phone)
    def allDoctors(self) :
        print('All Doctors:')
        print('Number of doctors:',self.num_doctor)
        print(self.doctor_dict)
    def allPatients(self) :
        print('All Patients;')
        print('Number of Patients:',self.num_patient)
        print(self.patient_dict)
    


    
h = Hospital("Evercare")
d1 = Doctor("1d","Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p","Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient ("2p","Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```

```python
# task 22 
class Vaccine :
    def __init__(self,vaccine_name,made_in,interval) :
        self.vaccine_name = vaccine_name 
        self.made_in = made_in 
        self.interval = interval 
class Person :
    def __init__(self,name,age,occupation='General Citizen') :
        self.name = name 
        self.age = age 
        self.occupation = occupation 
        self.vaccine_given = ''
        self.first_dose = False 
        self.second_dose = False 
    def pushVaccine(self,vaccine_name_given,dose='1st Dose') :
        if dose == '1st Dose' :
            if self.age >= 25 or self.occupation == 'Student' :
                self.vaccine_given= vaccine_name_given 
                self.first_dose = True 
                print('1st dose done for',self.name)
            else :
                print('Sorry',self.name,'Minimum age for taking vaccines is 25 years now')
        else:
            if self.vaccine_given.vaccine_name != vaccine_name_given.vaccine_name :
                print('Sorry',self.name,'you can\'t take 2 different vaccines')
            else :
                self.second_dose = True 
                print('2nd dose done for',self.name)
    def showDetail(self) :
        print('Name:',self.name,'Age:',self.age,'Type:',self.occupation)
        print('Vaccine name: ',self.vaccine_given.vaccine_name)
        if self.first_dose == True :
            print('1st dose: Given')
            print('2nd dose: Please come after',self.vaccine_given.interval,'days')
        elif self.second_dose == True : 
            print('1st dose: Given')
            print('2nd dose: Given')
    


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
